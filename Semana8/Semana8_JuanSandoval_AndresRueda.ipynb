```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
from mpl_toolkits.mplot3d import axes3d
from matplotlib import cm
import matplotlib as mpl
from tqdm import tqdm
```

```python
#Punto 1: Tomando el código del repositorio
Min, Max, N = 0.,40.,51
x = np.linspace(Min,Max,N)
y = x.copy()
h = x[1]-x[0]

def h1(y):
    return 100.
def h2(y):
    return 0.
def h3(x):
    return 0.
def h4(x):
    return 0.

def InitT():
    
    T = np.zeros((N,N))
    
    T[0,:] = h1(y)
    T[-1,:] = h2(y)
    
    T[:,0] = h3(x)
    T[:,-1] = h4(x)
    
    return T

def GetRelaxation(T, omega,Nit = int(1e5), tolerancia = 1e-2):
    
    itmax = 0
    
    for it in tqdm(range(Nit)):
        
        dmax = 0.
        
        for i in range(1, len(x)-1):
            for j in range(1, len(y)-1):
                tmp = 0.25*(T[i+1,j]+T[i-1,j]+T[i,j+1]+T[i,j-1])
                r = omega*(tmp - T[i,j])
                
                T[i,j] += r
                
                if np.abs(r) > dmax:
                    dmax = r
        
        if np.abs(dmax) < tolerancia:
            print(it)
            itmax = it
            break
            
    return T,itmax
```

```python
def generarDatos():
    omegas = np.linspace(1,1.9,10)
    its = []
    
    for i in range(len(omegas)):
        T = InitT()
        Tn,it =  GetRelaxation(T,omegas[i])
        its.append(it)
        print(it)
        
    return its

iteraciones = generarDatos()
```

```python
#Graficando
omegas = np.linspace(1,1.9,10)

figP1 = plt.figure(figsize=(5,5))
grP1 = figP1.add_subplot(111)
grP1.plot(omegas,iteraciones)
grP1.set_title("Iteraciones contra $\omega$ para la ecuación de Laplace")
grP1.set_xlabel("$\omega$")
grP1.set_ylabel("Iteraciones")
grP1.grid(True)

figP1.savefig("Semana8_Punto1.png")
```

```python
#Punto 4
#Condiciones iniciales y preliminares
v0 = 1
nu = 0.2
h = 0.02
R = v0*h/nu

yMin = 5
yMax = 30
xMin = 5
xMax = 25

N = 51
x = np.linspace(0,1,N)
y = np.linspace(0,1,N)
X,Y = np.meshgrid(x,y)

U = np.zeros((N,N))
W = np.zeros((N,N))

#Métodos para las condiciones de frontera
def vorticidad(U,W):
    
    #Bajando sobre x
    for i in range(xMin,xMax+1):
        W[i,0]= -2*(U[i,1] - U[i,0])/(h**2) #Izquierda
        W[i,-1] = -2*(U[i,-2] - U[i,-1])/(h**2) #Derecha
        
    #Yendo sobre y
    for j in range(N-yMax,N-yMin+1):
        W[0,j]= -2*(U[1,j] - U[0,j])/(h**2) #Arriba
        W[-1,j] = -2*(U[-2,j] - U[-1,j])/(h**2) #Abajo
        
def fronteras(U,W):
    for i in range(yMin,yMax+1):
        for j in range(xMin,xMax+1):
            U[xMin,j], U[xMax,j] = 0,0
            U[i,yMin], U[i,yMax] = 0,0
    
def velocidad(U,W):
    #Sobre todo el volumen
    for i in range(len(x)):
        U[-1,i] = U[-2,i] + v0*h*y[-1] #Método de Euler para la pared izquierda
        U[0,i] = U[1,i]
        U[i,0] = U[i,1]
        U[i,-1] = U[i,-2]
        
#Adaptado del método dado en clase
def solucion(Nit = 10000, omega = 0.8, tol=1e-5):
    for it in range(Nit):
        velocidad(U,W)
        vorticidad(U,W)
        fronteras(U,W)
        
        dmax1 = 0
        dmax2 = 0
        
        #Solucionando para U
        for i in range(1,len(x)-1):
            for j in range(1,len(y)-1):
                
                if (yMax +1>i > yMin+1) and (xMax +1> j > xMin+1):
                    U[i,j], W[i,j] = 0,0
                    
                else:
                    up = 0.25*(U[i+1,j]+U[i-1,j]+U[i,j+1]+U[i,j-1] + W[i,j]*h**2)
                    ru = omega*(up - U[i,j])
                    U[i,j]+=ru
                    
                    if np.abs(ru) > dmax1:
                        dmax1 = ru
        
        #Solucionando para W
        for i in range(1,len(x)-1):
            for j in range(1,len(y)-1):
                
                if (xMax +1>i > xMin+1) and (yMax +1> j > yMin+1):
                    U[i,j], W[i,j] = 0,0
                    
                else:
                    wp1 = 0.25*(W[i,j+1]+W[i,j-1]+W[i-1,j]+W[i+1,j])
                    wp2 = -(R/16)*(U[i,j+1]-U[i,j-1])*(W[i+1,j]-W[i-1,j])
                    wp3 = (R/16)*(W[i,j+1]-W[i,j-1])*(U[i+1,j]-U[i-1,j])
                    
                    rw = omega*(wp1+wp2+wp3-W[i,j])
                    W[i,j] += rw
                    
                    if np.abs(rw) > dmax2:
                        dmax2 = rw
                
        if (np.abs(dmax1) < tol) and (np.abs(dmax2) < tol):
            print(it)
            break
                
                
solucion()
```
